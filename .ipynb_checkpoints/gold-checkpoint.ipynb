### Data Processing

```python
df.columns=['date','price']
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.set_index('date',inplace=True)
```

```python
df.info()
```

```python
df.head()
```


### Data Analysis


```python
df.plot(figsize=(20,10)) 
plt.title('The Gold Monthly Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
```

```python
df['price'].describe().round(2)
```

```python
df[['price']].rolling(10).mean().plot(figsize=(20,10))
plt.title('The Gold Monthly Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
```

```python
_,ax=plt.subplots(figsize=(35,20))
sns.boxplot(x=df.index.year,y=df.price)
 
```
